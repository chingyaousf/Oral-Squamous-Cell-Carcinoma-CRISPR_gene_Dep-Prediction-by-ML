```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the dataset from the CSV file
file_path = '/content/drive/MyDrive/Colab Notebooks/DeepDEP/RF_error_values.csv'
df = pd.read_csv(file_path)

# Create a list of dataframes for MSE and MAE
dataframes_mse = [df[df['Gene'] == 'EIF2AK3']['MSE'],
                  df[df['Gene'] == 'MYC']['MSE'],
                  df[df['Gene'] == 'NQO1']['MSE'],
                  df[df['Gene'] == 'SF3B1']['MSE'],
                  df[df['Gene'] == 'SQSTM1']['MSE']]

dataframes_mae = [df[df['Gene'] == 'EIF2AK3']['MAE'],
                  df[df['Gene'] == 'MYC']['MAE'],
                  df[df['Gene'] == 'NQO1']['MAE'],
                  df[df['Gene'] == 'SF3B1']['MAE'],
                  df[df['Gene'] == 'SQSTM1']['MAE']]

# Box Plot
plt.figure(figsize=(10, 6))
plt.boxplot(dataframes_mse, positions=[0.9, 1.9, 2.9, 3.9, 4.9], widths=0.2, labels=['EIF2AK3', 'MYC', 'NQO1', 'SF3B1', 'SQSTM1'])
plt.boxplot(dataframes_mae, positions=[1.1, 2.1, 3.1, 4.1, 5.1], widths=0.2)
plt.xlabel('Gene')
plt.ylabel('Metric Value')
plt.title('MSE and MAE Distribution by Gene')
plt.ylim(0, 0.07)  # Adjust the y-axis range
plt.xticks([1, 2, 3, 4, 5], ['EIF2AK3', 'MYC', 'NQO1', 'SF3B1', 'SQSTM1'])
plt.legend(['MSE', 'MAE'])
plt.show()
```
